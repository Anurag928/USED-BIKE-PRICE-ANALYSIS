```python
Company=[]
Model=[]
kms=[]
cc=[]
owner=[]
Area=[]
SellingPrice=[]
CostPrice=[]


for i in range(1,25):
    url=f'https://www.credr.com/all-used-bikes?page={i}'
    driver.get(url)
    soup=BeautifulSoup(driver.page_source,'html.parser')
    
    for i in soup.find_all('div',class_='mb-3 bikeCard col-sm-6 col-md-6 col-lg-4'):
        x=i.find('h3',class_='font-semibold font16 bikeHdMarginBtm').text
        if(x):
            regex=re.findall('^(\w+)',x)
            if(regex):
                Company.append(regex[0])
            else:
                Company.append(np.nan)

            regex=re.findall('^\w+\s(.+)',x)
            if(regex):
                Model.append(regex[0])
            else:
                Model.append(np.nan)

        x=i.find('div',class_='card-body').text
        regex=re.findall('(\d+)\skms',x)
        if(regex):
            kms.append(regex[0])
        else:
            kms.append('0')

        regex=re.findall('(\d+)\scc',x)
        if(regex):
            cc.append(regex[0])
        else:
            cc.append(np.nan)

        regex=re.findall('(\d{1})\sowner',x)
        if(regex):
            owner.append(regex[0])
        else:
            owner.append(np.nan)


        x=i.find('div',class_='col-sm-12 col-12 text-muted bikeListItemsCustom').text
        if(x):
            Area.append(x)
        else:
            Area.append(np.nan)


        x=i.find('span',class_='greenColor').text
        if(x):
            SellingPrice.append(x)
        else:
            SellingPrice.append(np.nan)   


        x=i.find('span',class_='font14 font-regular pl-2 priceStrike')
        if(x):
            CostPrice.append(x.text)
        else:
            CostPrice.append(np.nan)
    
    
```
